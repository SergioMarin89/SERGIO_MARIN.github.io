```python
# Importar las librerias
import requests
from bs4 import BeautifulSoup
```

```python
# Solicitar acceso a la pagina web
page = requests.get("https://footballpredictions.com/footballpredictions/championsleaguepredictions/")

# crear el objeto soup
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
# Buscar el id correspondiente a los seven days forecast
buscar_clase = soup.find(class_="boxed-content-wrapper clearfix")
```

```python
print(type(buscar_clase))
```

```python
buscar_predicciones = buscar_clase.select('.bp-entry .prediction .predictionbox')

obtener_predicciones = [pt.get_text() for pt in buscar_predicciones]

print(len(obtener_predicciones))
```

```python
# crear lista de listas para poder limpiar la data
predicciones_lofl = [i for i in obtener_predicciones]

scores = []

for i,text in enumerate(predicciones_lofl):
    x = list(text.split(" "))
    scores.append(x[4])
    
scores
```

```python
buscar_equipos = buscar_clase.select('.bp-entry .prediction .predictionteams')

obtener_equipos = [txt.get_text() for txt in buscar_equipos]

print(obtener_equipos)
```

```python
buscar_texto_predicciones = buscar_clase.select('.bp-entry .prediction .bp-details')

obtener_texto_predicciones = [txt.get_text() for txt in buscar_texto_predicciones]

print(obtener_texto_predicciones)
```

```python
import pandas as pd
UEFA_CL = pd.DataFrame({
        "Equipos": obtener_equipos, 
        "Prediccion": scores, 
        "Analisis": obtener_texto_predicciones, 
    })
UEFA_CL
```

```python
UEFA_CL['Reales'] = ['4-2','2-2','1-2','5-2','0-0','1-3','3-0','1-2','4-1','3-0','0-3','1-2','1-3','3-0','1-0','1-2']
```

```python
UEFA_CL
```

```python
import numpy as np

UEFA_CL['ACERTO'] = np.where(UEFA_CL['Prediccion'] == UEFA_CL['Reales'], 'SI', 'NO')
```

```python
UEFA_CL
```

```python
Efectividad = (UEFA_CL['ACERTO']=='SI').sum()/len(UEFA_CL)

print('La Efectividad del pronóstico fue del: ' + '\033[95m'+'%.2f' %(Efectividad*100)+'%'+'\033[0m')
```
