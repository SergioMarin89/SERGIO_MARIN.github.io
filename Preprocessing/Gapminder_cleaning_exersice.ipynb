# Cleaning Data - Gapminder Dataset.

```python
import pandas as pd #Librería para manejo de la información en Data Frames.
import matplotlib.pyplot as plt #Librería para graficar
import numpy as np
```

```python
url = 'https://assets.datacamp.com/production/course_2023/datasets/gapminder.csv'

df = pd.read_csv(url, sep = ',', index_col=False)
```

```python
df.columns
```

```python
df = df.drop(['Unnamed: 0'], axis=1)
```

```python
df.info()
```

```python
df.describe()
```

Ya tenemos el DataFrame en su primera fase listo, esta disponible con las columnas correctas para ser analizado, ahora vamos a estudiar como ha crecido la población entre 1800 y 1899, el primer siglo de info que tenemos.

```python
df.plot(kind='scatter', x='1800', y='1899') # Se crea el scatter plot entre los datos de dos columnas
plt.xlabel('Expectativa de vida en 1800 por país') # Se crea la etiqueta del eje X
plt.ylabel('Expectativa de vida en 1899 por país') # Se crea la etiqueta del eje y
plt.xlim(20,55) # Se establecen los limites del eje x
plt.ylim(20,55) # Se establecen los limites del eje y
plt.show()
```

La grafica nos muestra un error en la información ya que los paises que estan en la línea diagonal quiere decir que su expectativa de vida no cambio con el avance del tiempo, puede ser que tengan el mismo dato durante todo el siglo y no en realidad el dato que corresponde a la realidad.

Al verificar con el método .sample() se puede generar muestras aleatorias de los datos y se confirma que efectivamente muchos países tienen el mismo dato desde 1800.

```python
df.sample(10)
```

```python
# Para continuar debemos reordenar las columnas, de modo que 'Life expectancy' sea la primer columna.

cols = df.columns.tolist() # Con .tolist() sacamos la lista de nombre de columnas 
cols = cols[-1:] + cols[:-1] # Con este código reordenamos, traemos el último al principio.
df = df[cols] 
df.head()
```

```python
def check_null_or_valid(row_data):
    """Function that takes a row of data,
    drops all missing values,
    and checks if all remaining values are greater than or equal to 0
    """
    no_na = row_data.dropna()[1:-1]
    numeric = pd.to_numeric(no_na)
    ge0 = numeric >= 0
    return ge0

# Check whether the first column is 'Life expectancy'
assert df.columns[0] == 'Life expectancy'

# Check whether the values in the row are valid
assert df.iloc[:, 1:].apply(check_null_or_valid, axis=1).all().all()

# Check that there is only one instance of each country
assert df['Life expectancy'].value_counts()[0] == 1
```

```python
df['Life expectancy'].value_counts()
```

Tenemos 3 filas por cada país, una fila contiene los datos desde 1800-1899, otra fila desde 1900-1999 y otra desde el año 2000. Debemos separarlos y hacer un solo dataset con una fila por país.

con el comando np.split() podemos separar los 3 dataframes y luego concatenarlos por las filas.

```python
paises = list(df['Life expectancy'])
```

```python
dfs = np.split(df,[0,101],axis=1)
```

```python
df1800s = pd.DataFrame(dfs[1])
df1800s.head()
```

```python
dfs1 = np.split(dfs[2],[0,100],axis=1)
```

```python
df1900s = pd.DataFrame(dfs1[1])
df1900s['Life expectancy'] = paises
cols1 = df1900s.columns.tolist() # Con .tolist() sacamos la lista de nombre de columnas 
cols1 = cols1[-1:] + cols1[:-1] # Con este código reordenamos, traemos el último al principio.
df1900s = df1900s[cols1] 
df1900s.head()
```

```python
df2000s = pd.DataFrame(dfs1[2])
df2000s['Life expectancy'] = paises
cols2 = df2000s.columns.tolist() # Con .tolist() sacamos la lista de nombre de columnas 
cols2 = cols2[-1:] + cols2[:-1] # Con este código reordenamos, traemos el último al principio.
df2000s = df2000s[cols2] 
df2000s.head()
```

```python
print(df1800s.shape)
print(df1900s.shape)
print(df2000s.shape)
```

```python
df1800s = df1800s.dropna()
df1900s = df1900s.dropna()
df2000s = df2000s.dropna()
print(df1800s.shape)
print(df1900s.shape)
print(df2000s.shape)
```

```python
# Merge los dataframes.

# Se pueden unir dataframe por varios métodos, uno de ellos es cancat y merging
# Concatenar une los dataframe tanto en filas como en colunmas, es decir, pega las 
# columnas pero adiciona las filas así esa fila ya existiera.

# Merge es mas parecido a un Join en SQL, junta las columnas pero no duplica las filas si se le
# indica mediante parámetros. 
# https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html?highlight=merge#pandas.DataFrame.merge

ndf = pd.merge(right=df1900s, right_on = df1900s.index.name, left = df1800s, left_on = df1800s.index.name, on=None )
gapminder= pd.merge(left = ndf, left_on = ndf.index.name, right=df2000s, right_on = df2000s.index.name, on=None)
```

```python
gapminder.sample(10)
```

```python
# volvemos a realizar los chequeos y esta vez no emite error, es decir, las 3 verificaciones estan ok.

# Check whether the first column is 'Life expectancy'
assert gapminder.columns[0] == 'Life expectancy'

# Check whether the values in the row are valid
assert gapminder.iloc[:, 1:].apply(check_null_or_valid, axis=1).all().all()

# Check that there is only one instance of each country
assert gapminder['Life expectancy'].value_counts()[0] == 1
```

## Organizar la data.

Ahora se debe organizar la data para que cada año sea una observación y para que el valor de Life expectancy sea la variable.


```python
gapminder_organziado = gapminder.melt(id_vars = 'Life expectancy')
gapminder.shape
```

```python
gapminder_organziado.head()
```

```python
gapminder_organziado.tail()
```

```python
gapminder_organziado.columns = ['Pais','Year','Expectativa_vida']
```

```python
gapminder_organziado.head()
```

```python
gapminder_organziado['Year'] = gapminder_organziado['Year'].astype(str)
gapminder_organziado['Expectativa_vida'] = pd.to_numeric(gapminder_organziado['Expectativa_vida'], errors = 'coerce')
```

```python
gapminder_organziado.head()
```

## Revisar el formato del texto de cada país, que este bien escrito.

Se procede a revisar el texto de cada país para asegurar uniformidad en el texto,

Se asegura que:

1. Los nombres inicen con Mayúscula.
2. Que exista un espacio entre palabras en nombres compuestos.
3. Que las abreviaturas tengan el punto al final de esta.

```python
# Se crea la lista paises con los nombres a evaluar
paises = gapminder_organziado['Pais']
paises = paises.drop_duplicates()
```

```python
patron = '^[A-Za-z\.\n]*$'
```

```python
mascara = paises.str.contains(patron) # Se revisa país por país si se tiene el patron
# Retorna objeto boleano.
```

```python
mascara
```

```python
mascara_falso = ~mascara #Para invertir el objeto boleano, true son lo que no coinciden.
mascara_falso
```

```python
error_paises = paises.loc[mascara_falso]
error_paises
```
